#### 2.2 Sample Joint Set Up 
<a id="fun_sample"></a>

```python
def sample_joint(age, p_diabetes, p_hyp):
    '''
    INPUT:
        - age: np.array of length n indicating the sampled age for each of the agents
        - p_diabetes: np.array of length 101 (number of ages) reporting the diabetes prevalence for each of the ages
        - p_hyp: np.array of length 101 (number of ages) reporting the diabetes prevalence for each of the ages
    OUTPUT: 
        - diabetes_status: np.array of length equal to n, made of 1 or 0 indicating whether each of the agents have or do not have diabetes
        - hyp_status: np.array of length equal to n, made of 1 or 0 indicating whether each of the agents have or do not have hypertension
    '''
    # Source: https://pubmed.ncbi.nlm.nih.gov/28701739/
    # (1) get the p of hypertension GIVEN DIABETES = 0.5
    # p(H|D) = p(H and D)/p(D)
    p_hyp_given_diabetes = 0.5
    
    # (2) get the p of hypertension GIVEN NOT DIABETES 
    # p(H|notD) = p(H and notD)/p(notD) = (p(H)-p(H and D))/ (1-p(D)) == (p(H)-p(H|D)*p(D))/ (1-p(D))
    p_hyp_given_not_diabetes = (p_hyp - p_hyp_given_diabetes*p_diabetes)/(1 - p_diabetes)
    
    # (3) retrieve the diabetes status using the p_diabetes 
    # Method for each agent (to be done for each of the generated agents with the associated age level etc): 
    #        - consider the agent age = age_i
    #        - consider the p_diabetes for age_i of the agent for the country of interest
    #        - sample a random number from a uniform distribution (0,1)
    #        - if random number < p_diabetes >> The agent has diabetes: TRUE (=1); otherwise it has not diabetes: FALSE (=0)
    diabetes_status = (np.random.rand(age.shape[0]) < p_diabetes[age]).astype(np.int)
    # (4) retrieve hypertension status 
    hyp_status = np.zeros(age.shape[0], dtype=np.int) # initialize a zeros-array to store hypertension statuses
    # (4.a) Count the number of individuals that have diabetes (diabetes_status == 1), sample random numbers and compare to p_hyp_given_diab
    hyp_status[diabetes_status == 1] = np.random.rand((diabetes_status == 1).sum()) < p_hyp_given_diabetes
    # (4.b) Count the number of individuals that do not have diabetes (diabetes_status == 0), sample random numbers and compare to p_hyp_given_not_diab
    hyp_status[diabetes_status == 0] = np.random.rand((diabetes_status == 0).sum()) < p_hyp_given_not_diabetes[age[diabetes_status == 0]]
    return diabetes_status, hyp_status
```

#### 2.3 Sample Joint Comorbidities Statuses
<a id="fun_joint"></a>

```python
def sample_joint_comorbidities(age, country):
    """
    INPUT: 
    - age: np.array of n elements indicating the ages of the sampled agents
    - country: pass string value for country from {'Italy','Spain','Germany','France'}; 
    OUTPUT: 
    - diabetes_status: np.array of length equal to n made of 1 or 0 indicating whether agents have or do not have diabetes
    - hyp_status: np.array of length equal to n made of 1 or 0 indicating whether agents have or do not have hypertension
    """
    return sample_joint(age, p_comorbidity(country, 'diabetes'), p_comorbidity(country, 'hypertension'))
```
